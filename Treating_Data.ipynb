```python
import pandas as pd
```

```python
PUZZLEPATH = ''
db = pd.read_csv(PUZZLEPATH)
```

```python
db
```

```python
mates = db.loc[db.Themes.str.contains('mate')].copy()
mates['Number'] = mates['Themes'].str.extract(r'(mateIn\d+)', expand=False)
mates = mates.drop(columns=['PuzzleId', 'Rating', 'RatingDeviation', 'Popularity', 'NbPlays', 'GameUrl', 'OpeningTags', 'Themes'])
mates['Number'] = mates['Number'].str.replace('mateIn', '')
mates
```

```python
notmates = db.loc[~db.Themes.str.contains('mate')].copy()
notmates = notmates.sample(n=300000)
notmates['Number'] = 0
notmates = notmates.drop(columns=['PuzzleId', 'Rating', 'RatingDeviation', 'Popularity', 'NbPlays', 'GameUrl', 'OpeningTags', 'Themes'])
notmates
```

```python
testing_dfs = []

# Sample 10% from matein1 to matein5
for i in range(1, 6):
    mate_in_i = mates[mates['Number'] == str(i)]
    testing_dfs.append(mate_in_i.sample(frac=0.1, random_state=42)) # Added random_state for reproducibility

# Sample 10% from notmates
testing_dfs.append(notmates.sample(frac=0.1, random_state=42)) # Added random_state for reproducibility

# Concatenate all parts into the testing dataframe
testing = pd.concat(testing_dfs)

# Display the new dataframe info and head
print(testing.info())
print(testing.head())
print(testing['Number'].value_counts())
```

```python
training = pd.concat([mates, notmates])
training = training[~training.index.isin(testing.index)]
training
```

```python
training.to_csv('trainingpuzzles.csv', index=False)
testing.to_csv('testingpuzzles.csv', index=False)
```
